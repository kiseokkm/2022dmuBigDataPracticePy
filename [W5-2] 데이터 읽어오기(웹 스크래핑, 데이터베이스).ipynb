```python
# 2) read_html(): 웹 페이지 내에 있는 테이블 형태의 데이터를 모두 반환하는 함수

data = pd.read_html('https://finance.naver.com/sise/sise_market_sum.nhn', encoding = 'cp949')
df = data[1]
df = df[df.종목명.notnull()] # 종목명이 NaN인 경우에는 제거
df
```

```python
df = df[df.종목명.notnull()] # 종목명이 NaN인 경우에는 제거
df
```

```python
# 실습. 10페이지 데이터 읽어서, 코스피 기업 500개 데이터 읽어오기

코스피500 = []
url = 'https://finance.naver.com/sise/sise_market_sum.nhn?&page='

for num in range(1, 11):
    full_url = url + str(num)
    #print(full_url)
    dfs = pd.read_html(full_url, encoding = 'CP949')
    df = dfs[1]
    df = df[df.종목명.notnull()]
    코스피500.append(df)
```

```python
코스피500[9]
```

```python
data = pd.concat(코스피500, axis = 0) # --> 다음 차시에 더 자세히 배움
```

```python
# 결과 데이터를 html 파일로 내보내기
data.to_html('data/코스피500.html', encoding = 'CP949')
```

```python
# 1. selenium 설치
# conda install selenium

from selenium import webdriver
```

```python
# 2. chromedriver 다운로드하여 적절한 위치에 옮기기
# 다운로드 사이트: https://sites.google.com/a/chromium.org/chromedriver/downloads
# 자신의 크롬 브라우저의 버전에 맞는 파일을 다운로드
```

```python
driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(3)
```

```python
driver.get(삼성전자url) # selenium이 해당 웹페이지를 실행
삼성_일별 = driver.page_source
```

```python
삼성주식 = pd.read_html(삼성_일별, encoding = 'cp949')
```

```python
삼성주식 = 삼성주식[0][삼성주식[0].날짜.notnull()]
```

```python
# 실습. 모든 기간의 삼성전자 주식 데이터를 스크래핑하여 삼성전자_일별주가.xlsx로 저장
삼성전자url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page='
일별주가 = []
for i in range(1, 619):
    full_url = 삼성전자url + str(i)
    driver.get(full_url) 
    삼성_일별 = driver.page_source
    삼성주식 = pd.read_html(삼성_일별, encoding = 'cp949')
    삼성주식 = 삼성주식[0][삼성주식[0].날짜.notnull()]
    일별주가.append(삼성주식)
    
삼성전자_일별주가 = pd.concat(일별주가, axis = 0)
```

```python
삼성전자_일별주가.to_excel('삼성전자_일별주가.xlsx')
```

## 4. 데이터베이스에서 읽기
### 파이썬에서 sqlite를 내장하고 있음

```python
import sqlite3
```

```python
#테이블 생성 SQL 정의
테이블생성 = 'CREATE TABLE locations \
                (a varchar[20], b varchar[20], c integer)'
```

```python
#sqlite 접속
conn = sqlite3.connect('temp.db')
```

```python
#테이블 생성 SQL 수행 
conn.execute(테이블생성)
```

```python
#conn.execute('drop table locations')
```

```python
# commit() 수행. - SQL 수행 후, commit을 해야만, 데이터베이스에 SQL 구문이 적용됨 
conn.commit()
```

```python
#생성된 테이블에 저장할 데이터 정의
data = [['서울','강남', 15]
       ,['서울','강동', 10]
       ,['서울','동대문', 7]
       ,['서울','양천', 12]
       ,['성남','분당', 12]
       ,['수원','영통', 15]
       ,['성남','중원', 11]
       ,['수원','팔달', 14]
       ]
```

```python
# 데이터 적재 SQL 정의 
적재 = 'insert into locations values (?,?,?)'
```

```python
#데이터 적재 SQL 수행
conn.executemany(적재, data)
```

```python
# commit() 수행
conn.commit()
```

```python
#Select SQL 실행1
pd.read_sql('SELECT * FROM locations', conn)
```

```python
#Select SQL 실행2
pd.read_sql('SELECT sum(c) FROM locations \
            GROUP BY a', conn)
```

```python
# 데이터베이스의 테이블 크기가 큰 경우에는 cursor 사용하기
cursor = conn.execute('SELECT * FROM locations')
```
