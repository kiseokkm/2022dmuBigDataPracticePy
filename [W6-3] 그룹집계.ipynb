```python
# 실습 데이터 적재
data = pd.read_excel('data/인구수예제.xlsx')
data.head(50)
```

```python
# 연도별 전체 총인구수 구하기
data.groupby('연도')['총인구'].sum()
```

```python
# 연도별 전체 총인구수 구하기
# SQL 구문으로 표현하면 아래와 같다.
# SELECT 연도, sum(총인구) 
# FROM data
# GROUP BY 연도
```

```python
data
```

```python
# 연도별 총인구수 구하기
data.groupby('연도')['총인구'].sum()
```

```python
#실습. 위 결과를 pivot_table()함수를 사용하여 그대로 출력 (pivot_table 배운 후)
data.pivot_table(index = '연도', aggfunc = 'sum', values = '총인구')
```

```python
# 연도별 전체의 남자인구, 여자인구, 총인구 수 구하기
data.groupby('연도')[['남자인구','여자인구','총인구']].sum()
```

```python
# 자치구별로 평균 총인구수 구하기
data.groupby('자치구')['총인구'].sum()
```

```python
# 도시/자치구별 평균 총인구수 구하기
data.groupby(['도시','자치구'])['총인구'].mean()
```

```python
# 실습 1. 도시/자치구별 평균 남자인구와 여자인구수 구하기
```

```python
data.groupby(['도시','자치구'])[['남자인구','여자인구']].mean()
```

```python
# 실습 2. 도시별, 연도별로 총인구수 출력
data.groupby(['도시', '연도'])['총인구'].sum().unstack(0)
```

```python
# pivot_table 한 후에 실습 2를 pivit_table() 함수로 변경
data.pivot_table(index = '연도', columns = '도시', aggfunc = 'sum', values = '총인구')
```

```python
#연도별 총인구수의 합
#data.groupby('연도')['총인구'].sum()
data.pivot_table(index = '연도', 
                aggfunc = 'sum',
                values = '총인구')
```

```python
# 도시별 연도별 총인구수의 합
#data.groupby(['도시','연도'])['총인구'].sum().unstack('도시')
data.pivot_table(index = '도시', columns = '연도',
                aggfunc = 'sum', values = '총인구')
```

```python
# 실습. 자치구 별로 남자인구의 평균을 구한 후, 남자 인구가 가장 많은 도시 및 자치구 Top 5 찾기
data.pivot_table(index = ['도시','자치구'], aggfunc = 'mean', values = '남자인구').sort_values(by = '남자인구', ascending = False).head()
```

```python
# 실습. 여자가 남자보다 많은 도시/자치구 상위 3개 찾기 
data['남녀차이'] = data.여자인구 - data.남자인구
data.head()
```

```python
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = '남녀차이')\
    .sort_values(by = '남녀차이', ascending = False).head(3)
```
